# Mean Correlation

```python
import numpy as np

# Initialize a list to store the mean correlation for each matrix
mean_correlations = []

# Iterate through the list of correlation matrices
for corr_matrix in correlation_coefficients:
    # Calculate the mean correlation for the current matrix
    mean_corr = np.mean(corr_matrix.values)

    # Append the mean correlation to the list
    mean_correlations.append(mean_corr)

# 'mean_correlations' now contains the mean correlation for each matrix in 'correlation_coefficients'
```

```python
mean_correlations[0]
```

```python
import numpy as np

# Initialize a list to store the maximum eigenvalues for each matrix
maximum_eigenvalues = []

# Iterate through the list of correlation matrices
for corr_matrix in correlation_coefficients:
    # Compute the eigenvalues of the current matrix
    eigenvalues = np.linalg.eigvals(corr_matrix.values)

    # Find the maximum eigenvalue
    maximum_eigenvalue = max(eigenvalues)

    # Append the maximum eigenvalue to the list
    maximum_eigenvalues.append(maximum_eigenvalue)

# 'max_eigenvalues' now contains the maximum eigenvalue for each correlation matrix
```

```python
# Initialize a new list to store the scaled maximum eigenvalues
scaled_maximum_eigenvalues = []

# Scale the maximum eigenvalues to the range [0, 1]
for eigenvalue in maximum_eigenvalues:
    scaled_eigenvalue = eigenvalue / 1000
    scaled_maximum_eigenvalues.append(scaled_eigenvalue)
```

```python
scaled_maximum_eigenvalues[0]
```

```python
import matplotlib.pyplot as plt

# Assuming you have calculated 'mean_correlations' and 'scaled_maximum_eigenvalues'

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot the mean correlations
ax1.plot(mean_correlations, label='Mean Correlation', marker='.', color = 'dodgerblue',  linewidth=1)
ax1.set_title('Mean Correlation')
ax1.set_xlabel('Time')
ax1.set_ylabel('Mean Correlation')


ax1.grid(True)


ax1.legend()

# Plot the scaled maximum eigenvalues
ax2.plot(scaled_maximum_eigenvalues, label='Scaled Max Eigenvalue', marker='.', color='orange', linewidth=1)
ax2.set_title('Maximum Eigenvalues')
ax2.set_xlabel('Time')
ax2.set_ylabel('Max Eigenvalue')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
