```python
import yfinance as yf
import numpy as np
import pandas as pd
matrix = pd.read_csv("/content/drive/MyDrive/Stock Analysis/SP_500/sp500_drop0.8.csv")
```

```python
# define window size
window_size = 20

# divide dataframe into 20-size windows and store them in a list
windows = [matrix[i:i+window_size] for i in range(0, len(matrix), window_size)]

# display the windows
for i, window in enumerate(windows):
    print(f"Window {i+1}:")
    print(window)
    print("\n")
```

```python
import numpy as np

# define window size
window_size = 20

# calculate correlation matrix for each window
corr_matrices = []
for i in range(0, len(matrix), window_size):
    window = matrix.iloc[i:i+window_size]
    corr_matrix = window.corr()
    corr_matrix = corr_matrix.replace(np.nan, 0) #replace NaN values with 0
    corr_matrices.append(corr_matrix)

# display the correlation matrices
for i, corr_matrix in enumerate(corr_matrices):
    print(f"Correlation matrix for window {i+1}:")
    print(corr_matrix)
    print("\n")
```

```python
corr_matrices[0]
```

```python
# distribute the data into windows and store the last date of each window in a list
last_dates = []
for i in range(0, len(matrix), window_size):
    window = matrix.iloc[i:i+window_size]
    last_date = window.iloc[-1]['Date']
    last_dates.append(last_date)

print(last_dates)
```
