# Average degree

```python
import networkx as nx

# Assuming you have already created the 'adjacency_matrices' list

# Create a list to store the average degrees for each graph
average_degrees = []
# Create a list of node names (assuming your adjacency matrix has column and row names)
node_names = adjacency_matrices[0].columns.tolist()

# Iterate through the list of adjacency matrices
for adjacency_matrix in adjacency_matrices:
    # Create a graph from the adjacency matrix
    G = nx.Graph()
    G.add_nodes_from(node_names)

    for i in range(len(node_names)):
        for j in range(i + 1, len(node_names)):
            if adjacency_matrix.iloc[i, j] == 1:
                G.add_edge(node_names[i], node_names[j])

    # Calculate the average degree
    degrees = dict(G.degree())
    avg_degree = sum(degrees.values()) / len(degrees)
    average_degrees.append(avg_degree)

# Print the average degrees for each graph
for i, avg_degree in enumerate(average_degrees):
    print(f"Graph {i + 1}: Average Degree = {avg_degree:.2f}")
```
